```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import pandas as pd
```

```python
reviews_df = pd.read_csv(r"C:\Users\avinash\Downloads\Dataset  (1).csv")
```

```python
reviews_df
```

```python

stop_words = set(stopwords.words('english'))
restaurant_names_tokenized = []
for name in reviews_df['Restaurant Name']:
    tokens = word_tokenize(name.lower())
    filtered_tokens = [word for word in tokens if word not in stop_words]
    restaurant_names_tokenized.append(filtered_tokens)

positive_words = []
negative_words = []
for tokens in restaurant_names_tokenized:
    for token in tokens:
        if token.endswith('ing') or token.endswith('ed'):
            positive_words.append(token)
        elif token.startswith('un') or token.startswith('in'):
            negative_words.append(token)

positive_word_counts = Counter(positive_words)
negative_word_counts = Counter(negative_words)
print("Most common positive keywords:", positive_word_counts.most_common(5))
print("Most common negative keywords:", negative_word_counts.most_common(5))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reviews_df = pd.read_csv(r"C:\Users\avinash\Downloads\Dataset  (1).csv")
```

```python
reviews_df['city_length'] = reviews_df['City'].apply(len)

avg_city_length = reviews_df['city_length'].mean()
print(f"Average city name length: {avg_city_length:.2f} characters")

city_lengths_by_agg_rating = reviews_df.groupby('Aggregate rating')['city_length'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.pie(city_lengths_by_agg_rating['city_length'], labels=city_lengths_by_agg_rating['Aggregate rating'], autopct='%1.1f%%')
plt.title('Average City Name Length by Aggregate Rating')
plt.axis('equal')  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\avinash\Downloads\Dataset  (1).csv")
```

```python
data
```

```python
votes_by_restaurant = data.groupby('Restaurant Name')['Votes'].sum().reset_index()
votes_by_restaurant = votes_by_restaurant.sort_values(by='Votes', ascending=False)
print(f"Restaurant with the highest number of votes: {votes_by_restaurant.iloc[0]['Restaurant Name']} ({votes_by_restaurant.iloc[0]['Votes']} votes)")
non_zero_votes = votes_by_restaurant[votes_by_restaurant['Votes'] > 0]
print(f"Restaurant with the lowest number of votes: {non_zero_votes.iloc[-1]['Restaurant Name']} ({non_zero_votes.iloc[-1]['Votes']} votes)")
plt.figure(figsize=(10, 6))
plt.hist(votes_by_restaurant['Votes'], bins=20, edgecolor='black')
plt.xlabel('Number of Votes')
plt.ylabel('Count of Restaurants')
plt.title('Distribution of Votes across Restaurants')
plt.show()
```

```python

data = pd.read_csv(r"C:\Users\avinash\Downloads\Dataset  (1).csv")
```

```python
Price Range vs. Online Delivery and Table Booking
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv(r"C:\Users\avinash\Downloads\Dataset  (1).csv")

pivot_table = pd.pivot_table(data, values='Restaurant Name', index=['Price range', 'Has Online delivery'], columns='Has Table booking', aggfunc='count', fill_value=0)


pivot_table = pivot_table.reset_index()


pivot_table.columns = ['Price range', 'Online Delivery', 'Table Booking: No', 'Table Booking: Yes']


fig, ax = plt.subplots(figsize=(10, 6))
sns.set_style("whitegrid")

for price_range in pivot_table['Price range'].unique():
    subset = pivot_table[pivot_table['Price range'] == price_range]
    ax.bar(subset['Online Delivery'], subset['Table Booking: No'], color='lightblue', alpha=0.7, label=f'{price_range} - Table Booking: No')
    ax.bar(subset['Online Delivery'], subset['Table Booking: Yes'], bottom=subset['Table Booking: No'], color='orange', alpha=0.7, label=f'{price_range} - Table Booking: Yes')


ax.set_xticks([0, 1])
ax.set_xticklabels(['No Online Delivery', 'Online Delivery'])
plt.xticks(rotation=45)


ax.set_xlabel('Online Delivery')
ax.set_ylabel('Count of Restaurants')
ax.set_title('Relationship between Price Range, Online Delivery, and Table Booking')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\avinash\Downloads\Dataset  (1).csv")
```

```python
data
```

```python
data = data.dropna(subset=['Price range', 'Has Online delivery', 'Has Table booking'])
data['Price range'] = data['Price range'].astype('category')
data['Has Online delivery'] = data['Has Online delivery'].astype('bool')
data['Has Table booking'] = data['Has Table booking'].astype('bool')


price_range_map = {category: index for index, category in enumerate(data['Price range'].cat.categories)}
data['Price range Numeric'] = data['Price range'].map(price_range_map)


online_delivery_corr = data['Has Online delivery'].corr(data['Price range Numeric'], method='kendall')
table_booking_corr = data['Has Table booking'].corr(data['Price range Numeric'], method='kendall')

print(f"Correlation between Price Range and Online Delivery: {online_delivery_corr}")
print(f"Correlation between Price Range and Table Booking: {table_booking_corr}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(data=data, x='Price range', hue='Has Online delivery')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Online Delivery by Price Range')

plt.subplot(1, 2, 2)
sns.countplot(data=data, x='Price range', hue='Has Table booking')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Table Booking by Price Range')

plt.tight_layout()
plt.show()
```
